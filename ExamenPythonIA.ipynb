```python

#Escriba una función que muestre “n” cantidad de números de la sucesión de Fibonacci, siendo “n” un argumento numérico dentro de la función.

def fibonacci(n):
    secuencia = []
    a, b = 0, 1
    for _ in range(n):
        secuencia.append(a)
        a, b = b, a + b
    return secuencia

print(fibonacci(10))



```

```python
from collections import Counter

def valor_mas_repetido(lista_numeros):
    if not lista_numeros:
        return None
    contador = Counter(lista_numeros)
    valor_mas_comun = contador.most_common(1)[0][0]
    return valor_mas_comun

lista = [1, 3, 2, 1, 4, 1, 3, 2, 2, 2]
print(valor_mas_repetido(lista))
```
